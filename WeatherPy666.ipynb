```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from time import sleep

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
#output_data_file = "output_data/cities.csv"
#output_df = pd.read_csv("output_data/cities.csv") --> this didn't work

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs) #create variable to hold both lat and long

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city) 

#Print the city count to confirm sufficient count
#len(cities)
#cities
```

## Perform API Calls

```python
# Define your own API key
api_key = api_keys.api_key

# Start URL for Weather Map API Call. units=Imperial. and ADD this part: &q=
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key + "&q="

#set up the lists and variables to hold info.
#cities = cities #already established above
latitude = [] #adding variables to hold the request's info on latitude
temperature = [] #adding variable to hold the request's info on temp
humidity = []
wind = []
clouds = []
name = []
#set up counter for the processing log record counter
counter = 1 
running = True #this is used so that the script doesn't run forever - I problem I had in testing only 6 cities
# Loop through the list of cities and perform a request for data on each
while running:
    for i in range(len(cities)):  
        city = cities[i]
        response = requests.get(url + city).json()
        try:
            name.append(response['name'])
            latitude.append(response['coord']['lat'])
            temperature.append(response['main']['temp'])
            humidity.append(response['main']['humidity'])
            wind.append(response['wind']['speed'])
            clouds.append(response['clouds']['all'])
            print(f"Processing Record {counter} | {city}")
            print(f"{url}{city}")
            time.sleep(2) #this slows down the requests so that it doesn't exceed openweathermap's limits
            running = False
            counter = counter + 1
        except (KeyError, TimeoutError, ConnectionError): #this finds missing records and skips them
            cities.append(response['cod'])
            print("City not found. Skipping to next record.")
            time.sleep(5)
```

```python
#for creating scatters: refer to 2/Activites/ins_openWeatherdataframe
#1. create a dataframe from cities,latlng,temp,humidity, etc
#use previously established variables from above

weather_dict = {"City": cities,
                "Latitude": latitude,
                "Temperature": temperature,
                "Humidity": humidity,
                "Cloudiness": clouds,
                "Wind Speed": wind
               }

weather_df = pd.DataFrame.from_dict(weather_dict, orient='index')
#weather_df.drop(weather_df.filter(like='None').columns, 1)#weather_df.apply(pd.DataFrame.describe)

#weather_df.dropna()
weather_df.transpose()
#df = pd.DataFrame(weather_df.values.T,columns=weather_df.columns)
#df.head()
#weather_df = weather_df[weather_df['City'] !='404']
weather_df = weather_df.replace('404', np.nan)
#weather_df.columns.values
weather_df
#new_df = weather_df.dropna()
#new_df
weather_df.dropna(axis=1)
weather_df
```

```python
#2. build scatter for each data type. facecolors="red", alpha=0.75
plt.scatter(weather_df["Latitude"], weather_df["Temperature"], marker="o", alpha=0.75)
plt.title("Temperature in World Cities (July 2018)")
plt.ylabel("Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)
plt.savefig("TempInWorldCities.png")
plt.show()
```
